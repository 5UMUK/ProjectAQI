```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd

df = pd.read_csv("../Data/pollution_data.csv")
```

```python
# Check for null values and replace with means for each year
 
print(df.isnull().sum())
null_counts = df.isnull().sum()
for column in df.columns:
    if null_counts[column] > 0:
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
df.isnull().sum()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 6))
sns.lineplot(x='Year', y='PM2.5', data=df, marker='o')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.title('Year vs PM2.5')

# Show the plot
plt.show()
```

```python
# TODO: Remove PM2.5 vs PM2.5
var_ind = [2, 3, 4, 5, 6]
plt.figure(figsize=(12, 10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
for i in range(1, 6):
    plt.subplot(3, 3, i)
    sns.scatterplot(x=df.iloc[:, var_ind[i - 1]], y="PM2.5", data=df)
    sns.regplot(x=df.iloc[:, var_ind[i - 1]], y="PM2.5", data=df, scatter=False, color='blue')

    plt.title("PM2.5 vs. " + df.columns[var_ind[i - 1]])
```

```python
variables = ["CO", "NO2", "O3", "PM2.5", "SO2"]
for variable in variables:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="State", y=variable, data=df)
    plt.title(f"{variable} by States")
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Create a pairplot
sn.pairplot(df[["Year", "PM2.5", "CO", "NO2", "O3", "SO2"]])
plt.suptitle("Pairplot of PM2.5, CO, NO2, O3, SO2")
plt.show()
```

```python
var_ind = [2, 3, 4, 5, 6]
plt.figure(figsize=(12, 10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
for i in range(1, 6):
    plt.subplot(3, 3, i)
    # plt.scatter(x=df["Year"], y=df.iloc[:, var_ind[i - 1]])
    sn.lineplot(x=df["Year"], y=df.iloc[:, var_ind[i - 1]], data=df, marker='o')
    plt.title("Year vs. " + df.columns[var_ind[i - 1]])
```

```python
variables = ["CO", "NO2", "O3", "PM2.5", "SO2"]
for variable in variables:
    plt.figure(figsize=(10, 6))
    sn.boxplot(x="State", y=variable, data=df)
    plt.title(f"{variable} by States")
    plt.xticks(rotation=90)
    plt.show()
```
